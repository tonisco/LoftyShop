```python
def get_keybert_phrase(text):

    kw_model = KeyBERT()
    keywords_with_scores = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 3))

    keywords = [x[0] for x in keywords_with_scores]

    # Remove duplicate keywords with customized threshold
    # unique_keywords = list(process.dedupe(keywords, threshold=75))
    # extracted_keywords = ", ".join(unique_keywords[:top_N])
    extracted_keywords = ", ".join(keywords)

    return extracted_keywords
```

```python
def get_keybert_phrase_by_line(text):
  # text_list = text.split('\n')
  text_list = sent_tokenize(text)
  all_phrase = []
  for value in text_list:
    all_phrase.append(get_keybert_phrase(value))
  return ','.join(all_phrase)
```

```python
def get_YAKE_phrase(text):

    # specifying parameters
    language = "en"
    max_ngram_size = 3
    # deduplication_threshold = 0.75
    # deduplication_algo = 'seqm'
    # windowSize = 1
    # numOfKeywords = top_N

    custom_kw_extractor = yake.KeywordExtractor(lan=language,
                                                n=max_ngram_size,
                                                # dedupLim=deduplication_threshold,
                                                # dedupFunc=deduplication_algo,
                                                # windowsSize=windowSize,
                                                # top=numOfKeywords,
                                                # features=None
                                                )

    keywords_with_scores = custom_kw_extractor.extract_keywords(text)
    keywords = [x[0] for x in keywords_with_scores]

    # Remove duplicate keywords with customized threshold
    # unique_keywords = list(process.dedupe(keywords, threshold=75))
    extracted_keywords = ", ".join(keywords)

    return extracted_keywords
```

```python
class _ContinuationToken:
    __slots__ = (
        "token",
        "lang",
        "country",
        "sort",
        "count",
        "filter_score_with",
        "filter_device_with",
    )

    def __init__(
        self, token, lang, country, sort, count, filter_score_with, filter_device_with
    ):
        self.token = token
        self.lang = lang
        self.country = country
        self.sort = sort
        self.count = count
        self.filter_score_with = filter_score_with
        self.filter_device_with = filter_device_with
```

```python
from google_play_scraper import reviews, Sort, app, search
import random
import time
import csv
import os

path = "/content/drive/MyDrive/"

class GooglePlay:
    _reviews = []
    _previous_review_ids = []

    def __init__(self, app_id: str, lang="en", country="us"):
        self._app_id = app_id
        self._lang = lang
        self._country = country
        self._title = ""
        self._details = {}

    def details(self):
        if self._title == "":
            app_details = app(app_id=self._app_id, lang=self._lang, country=self._country)
            self._title = app_details["title"]
            self._details = app_details
        return self._details

    def reviews(self):
        return self._reviews

    def fetch_review(
        self,
        sort: Sort = Sort.NEWEST,
        count: int = 200,
        _save = True,
        continuation_token: any = None,  # type: ignore
    ):
        assert count <= 200, "Count must be not be greater than 200"

        rvws, token = reviews(
            app_id=self._app_id,
            lang=self._lang,
            country=self._country,
            sort=sort,
            continuation_token=continuation_token,
            count=count,
        )

        if _save:
          self._reviews = rvws

        return rvws, token

    def fetch_all_reviews(
        self, sort: Sort = Sort.NEWEST, count: int = 200, batch_range=4999
    ):
        assert batch_range >= 2

        # current review batch
        batch_num = 0
        # number of times no new review was added
        empty_num = 0


        # check for previous token
        pre_token = self.read_token_csv()

        if pre_token and len(pre_token) > 0:
          last_token = pre_token[len(pre_token)-1]

          token = self.create_token(
              last_token["token"],
              last_token["lang"],
              last_token["country"],
              last_token["sort"],
              int(last_token["count"]),
              last_token["filter_score_with"],
              last_token["filter_device_with"]
              )

          # check for previous reviews
          pre_reviews = self.read_reviews_csv()
          if pre_reviews and len(pre_reviews) > 0:
            self._reviews = pre_reviews
            print(len(pre_reviews),len(self._reviews))

          rvws, token = self.fetch_review(
              sort=sort,
              count=count,
              continuation_token=token,
              _save=False
          )

          self._reviews.extend(rvws)
          print(len(pre_reviews),len(self._reviews))

          self.write_token_to_csv(token)
          self.write_reviews_to_csv(rvws)

          print(f"Batch taken from data")
        else:
          rvws, token = self.fetch_review(
              sort=sort,
              count=count,
              _save=False
          )

          self._reviews.extend(rvws)

          self.write_token_to_csv(token)
          self.write_reviews_to_csv(rvws)

          batch_num += 1

          print(f"Batch {batch_num} completed.")
          time.sleep(random.randint(1, 5))

        self._previous_review_ids = []
        for r in self._reviews:
            self._previous_review_ids.append(r["reviewId"])

        print(len(self._reviews))

        prev_token = token

        for batch in range(batch_range):
            rvws, token = self.fetch_review(
                sort=sort,
                count=count,
                continuation_token=token,
                _save=False,
            )

            # TODO: maybe use as default checker or remove
            print(prev_token == token.token)
            if prev_token.token != token.token:
                self.write_token_to_csv(token)
                prev_token = token

            new_review_ids = []
            selected_reviews = []
            print(len(self._reviews))

            for r in rvws:
                new_review_ids.append(r["reviewId"])
                # check if it has been added before
                if r["reviewId"] not in self._previous_review_ids:
                    # save to selected if it has not been added before
                    selected_reviews.append(r)
                # else:
                #     # print both values
                #     selected=next((x for x in self._reviews if x["reviewId"] == r["reviewId"]), None)
                #     print(r['reviewId'], selected['reviewId'])

            # Add the list of review dicts to main app_reviews list
            self._reviews.extend(selected_reviews)
            if len(selected_reviews) > 0:
              self.write_reviews_to_csv(selected_reviews)

            # Increase batch count by one
            batch_num += 1
            print(f"Batch {batch_num} completed.")
            print(len(self._reviews))
            # Break loop and stop scraping for current app if most recent batch
            # did not add any unique reviews
            print(len(new_review_ids), len(self._previous_review_ids))
            all_review_ids = self._previous_review_ids + new_review_ids
            print(len(self._previous_review_ids), len(all_review_ids))
            if len(set(self._previous_review_ids)) == len(set(all_review_ids)):
                if empty_num >= 3:
                    print(
                        f"No reviews left to scrape. Completed {batch_num} batches.\n"
                    )
                    break
                else:
                    empty_num += 1
            else:
                empty_num = 0

            self._previous_review_ids = all_review_ids

            # Wait 1 to 5 seconds to start next batch
            time.sleep(random.randint(1, 5))

    @staticmethod
    def search(search_string: str, lang="en", country="us"):
        return search(search_string, lang=lang, country=country)

    def create_token( self, token, lang, country, sort, count, filter_score_with, filter_device_with):
      return _ContinuationToken(token, lang, country, sort, count, filter_score_with, filter_device_with)

    def write_to_csv(self,name="",data:list=[]):
        assert name != "","no name provided"
        assert len(data) > 0, "no data provided"

        name = f"{name}.csv"

        if os.path.isfile(name) and os.stat(name).st_size != 0:
          with open(name, 'a') as file:
              writer = csv.writer(file)

              for review in data:
                writer.writerow(review.values())
        else:
            with open(name, 'w+') as file:
                writer = csv.writer(file)

                header = data[0].keys()
                writer.writerow(header)

                for review in data:
                  writer.writerow(review.values())

    def write_reviews_to_csv(self,reviews):
        assert self._title != "","No app title, fetch details"
        assert len(reviews) > 0

        name = f"{path}{self._title}"

        self.write_to_csv(name=name,data=reviews)

    def write_token_to_csv(self,token):
        assert self._title != "","No app title, fetch details"
        assert token != None

        name=f"{path}{self._title}_token"

        self.write_to_csv(name=name,data=[
            {
                "token" : token.token,
                "lang" : token.lang,
                "country" : token.country,
                "sort" : token.sort,
                "count" : token.count,
                "filter_score_with" : token.filter_score_with,
                "filter_device_with" : token.filter_device_with
            }
          ])

    def read_csv(self,name):
        assert self._title != "","No app title, fetch details"

        filename = f"{name}.csv"
        data = []
        if os.path.isfile(filename) and os.stat(filename).st_size != 0:
          with open(filename,'r+') as file:
            reader = csv.DictReader(file)
            for dic in reader:
              data.append(dic)
        return data

    def read_token_csv(self):
        assert self._title != "","No app title, fetch details"

        return self.read_csv(f"{path}{self._title}_token")

    def read_reviews_csv(self):
        assert self._title != "","No app title, fetch details"

        return self.read_csv(f"{path}{self._title}")
```

```python
def extract_information(data):
  try:
    output= {
        'target_analysis':{},
        'review':{},
    }

    # Sentiment analysis
    sentiment_score = sid.polarity_scores(data['content'])
    if sentiment_score['compound'] >= 0.05:
        output['target_analysis']['Sentiment Analysis'] = 'Positive'
    elif sentiment_score['compound'] <= -0.05:
        output['target_analysis']['Sentiment Analysis'] = 'Negative'
    else:
        output['target_analysis']['Sentiment Analysis'] = 'Neutral'


    # Topic modeling
    vectorizer = CountVectorizer(stop_words='english')
    X = vectorizer.fit_transform([data['content']])
    lda = LatentDirichletAllocation(n_components=1, random_state=42)
    lda.fit(X)
    feature_names = vectorizer.get_feature_names_out()
    output['target_analysis']['Topic Modeling']= feature_names[lda.components_[0].argmax()]

    # Key phrase extraction using RAKE
    r = Rake()
    r.extract_keywords_from_text(data['content'])
    output['target_analysis']['Key Phrases']= r.get_ranked_phrases()

    # Aspect extraction
    doc = nlp(data['content'])
    aspect = [ent.text for ent in doc.ents]
    output['target_analysis']['Aspect Extraction']= aspect

    output['review']['content'] = data['content']
    output['review']['created_at'] = data['at']
    output['review']['rate'] = data['score']
    output['review']['reviewer_name'] = data['userName']

    return output
  except Exception as error:
    print('error')
    print(error)
    return
```

```python
def nmf_preprocess(text):
    # Tokenize
    words = word_tokenize(text)
    # Remove stopwords
    words = [word for word in words if word.lower() not in stopwords.words('english')]
    return ' '.join(words)
```

```python
def nmf_analyze(text):
  text = nmf_preprocess(text)
  vectorizer = TfidfVectorizer(max_features=1500, min_df=10, stop_words='english')
  X = vectorizer.fit_transform([text])
  words = np.array(vectorizer.get_feature_names_out())

  nmf = NMF(n_components=10, solver="mu")
  W = nmf.fit_transform(X)
  H = nmf.components_

  # Transform the TF-IDF matrix using the NMF model
  # doc_topic_dist = nmf.transform(X)

  topic_array = []

  for i, topic in enumerate(H):
     topic_array.push("".join(',',[str(x) for x in words[topic.argsort()[-10:]]]))
```

```python
# Function to preprocess the text
def lda_preprocess(text):
    # Tokenize
    words = simple_preprocess(str(text), deacc=True)
    # Remove stopwords
    words = [word for word in words if word not in stopwords.words('english')]
    # Lemmatize
    doc = nlp(" ".join(words))
    return [token.lemma_ for token in doc]
```

```python
def lda_analyze(text):
  # Preprocess the text
  text = lda_preprocess(text)
  # Create a dictionary and corpus
  dictionary = corpora.Dictionary([text])
  corpus = [dictionary.doc2bow(text)]
  # Train the LDA model
  lda_model = LdaModel(corpus=corpus, num_topics=10, id2word=dictionary)

  topics_list = []
  # Print the topics
  for idx, topic in lda_model.print_topics(-1):
    topics_list.append(topic)
  return topics_list
```

## CRF (Conditional Random Fields)

```python
# prompt: Aspect Extraction with CRF (Conditional Random Fields)

import nltk
from nltk.corpus import conll2000
from nltk.tag.crf import CRFTagger

# Load the CoNLL 2000 dataset
train_data = conll2000.chunked_sents('train.txt', chunk_types=['NP'])
test_data = conll2000.chunked_sents('test.txt', chunk_types=['NP'])

# Train the CRF tagger
tagger = CRFTagger()
tagger.train(train_data, 'model.crf.tagger')

# Evaluate the tagger
accuracy = tagger.evaluate(test_data)
print('Accuracy:', accuracy)

# Extract aspects from a new sentence
sentence = 'I love the food at this restaurant.'
tokens = nltk.word_tokenize(sentence)
tagged_tokens = tagger.tag(tokens)

# Find the aspects
aspects = []
for token, tag in tagged_tokens:
    if tag == 'NP':
        aspects.append(token)

# Print the aspects
print('Aspects:', aspects)
```

```python
# Load spaCy NER model
nlp = spacy.load("en_core_web_sm")
```

```python
def spacy_analyze(text):
  doc = nlp(text)

  aspects = []

  for token in doc:
    # Identify noun chunks and entities as aspects
    if token.dep_ in ['nsubj', 'dobj'] and token.pos_ == 'NOUN':
        aspects.append(token.text)
    elif token.ent_type_ in ['ORG', 'PRODUCT', 'GPE', 'LOC']:
        aspects.append(token.text)

  return aspects
```
